```python
# Convert ONNX to TFLite
import os
from pathlib import Path

output_dir = "."
model_name = Path(onnx_file).stem

print(f"Converting {onnx_file} to TFLite...")
!onnx2tf -i {onnx_file} -o {output_dir} -osd

# Find the output file
tflite_file = f"{model_name}_float32.tflite"
if os.path.exists(tflite_file):
    print(f"‚úÖ Conversion successful!")
    print(f"TFLite file: {tflite_file}")
else:
    print("‚ö†Ô∏è  Output file not found, checking directory...")
    !ls -la *.tflite 2>/dev/null || echo "No TFLite files found"
```
